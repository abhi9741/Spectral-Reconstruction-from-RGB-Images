```python
import os 
import scipy.io as io 
import numpy as np 
import cv2 
import matplotlib.pyplot as plt
```

```python
sample_file = "sample/ARAD_HS_0464.mat"
sample_rgb = "sample/ARAD_HS_0464_RealWorld.jpg"
```

```python
sample_dump = io.loadmat(sample_file)
print(sample_dump.keys())
```

```python
img_hsi = sample_dump['cube'] #hyperspectral cube
img_bands = sample_dump['bands'][0] #bands and corresponding wavelengths
print(img_hsi.shape)
print(img_hsi.dtype)
```

```python
img_256 = (img_hsi*255).astype(np.uint8)
```

```python
print(img_hsi[0][0][0])
print(img_256[0][0][0])
```

```python
chnl = 1
img_chnl = img_256[:,:,chnl]
```

```python
img_save_path = sample_file.replace(".mat",f"_{chnl}.png")
```

```python
cv2.imwrite(img_save_path, img_chnl)
```

```python
print("channel selected :",img_bands[chnl-1])
```
